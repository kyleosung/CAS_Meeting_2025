```python
from src.utils import load_dataframe
from src.constants import DATA_CLEANED

df = load_dataframe(DATA_CLEANED)
df.head()
```

```python
df_non_zero_amount = df[df["amount"] != 0]
df_non_zero_amount.head()
```

```python
print("Claim Frequency by Coverage Type")
print("=" * 30)
print(
    f"Personal Property: {df_non_zero_amount[df_non_zero_amount['coverage'] == 'Personal Property'].shape[0]}"
)
print(
    f"Liability: {df_non_zero_amount[df_non_zero_amount['coverage'] == 'Liability'].shape[0]}"
)
print(
    f"Guest Medical: {df_non_zero_amount[df_non_zero_amount['coverage'] == 'Guest Medical'].shape[0]}"
)
print(
    f"Additional Living Expense: {df_non_zero_amount[df_non_zero_amount['coverage'] == 'Additional Living Expense'].shape[0]}"
)
```

```python
coverage_sum = df_non_zero_amount.groupby("coverage")["amount"].sum()
print("Total Claim Amount across Coverage Types")
print("=" * 40)
print(coverage_sum)
```

```python
from src.constants import DATA_DIR

personal_property_df = df[df["coverage"] == "Personal Property"]
personal_property_df = personal_property_df.drop(columns=["coverage"])
personal_property_df.to_csv(DATA_DIR / "data_cleaned_PersonalProperty.csv", index=False)
personal_property_df.head()
```

```python
guest_medical_df = df[df["coverage"] == "Guest Medical"]
guest_medical_df = guest_medical_df.drop(columns=["coverage"])
guest_medical_df.to_csv(DATA_DIR / "data_cleaned_Guest_medical.csv", index=False)

liability_df = df[df["coverage"] == "Liability"]
liability_df = liability_df.drop(columns=["coverage"])
liability_df.to_csv(DATA_DIR / "data_cleaned_liability.csv", index=False)

add_living_exp_df = df[df["coverage"] == "Additional Living Expense"]
add_living_exp_df = add_living_exp_df.drop(columns=["coverage"])
add_living_exp_df.to_csv(
    DATA_DIR / "data_cleaned_AdditionalLivingExpense.csv", index=False
)
```

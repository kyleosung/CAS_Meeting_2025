```python
from src.utils import load_dataframe
from src.constants import DATA_DIR, IMG_DIR

import matplotlib.pyplot as plt
```

```python
datasets = ["PersonalProperty", "AdditionalLivingExpense", "Guest_medical", "liability"]

for dataset in datasets:
    for label, title in [("amount", "Claim Amount"), ("risk_tier", "Risk Tier")]:
        df = load_dataframe(DATA_DIR / f"data_cleaned_{dataset}.csv")

        average_claim_amounts = df.groupby("class")[label].mean()

        plt.plot(average_claim_amounts)
        plt.xlabel("Class")
        plt.ylabel(f"Average {title}")
        plt.title(f"{dataset}: Average {title}")
        plt.savefig(IMG_DIR / f"{dataset}_{label}.png")
        plt.show()
```
